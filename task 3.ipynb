## Exploratory Data analysis

```python
data.plot(x='Sales', y='Profit', style='o')  
plt.title('Sales vs Profit')  
plt.xlabel('Sales')  
plt.ylabel('Profit')  
plt.show()
```

```python
sns.stripplot(x='Discount', y='Profit', data=data)
plt.title('Graph')
 
# function to show plot
plt.show()
```

```python
data.hist(bins=50 ,figsize=(16,13))
plt.show();
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Category', data=data)
plt.show()
```

```python
plt.figure(figsize=(17,6))
sns.countplot(x='Sub-Category', data=data)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Segment', data=data)
plt.show()
```

## plotting a graph for Quantity VS Sales data

```python
sns.set(style="whitegrid")
plt.figure(2, figsize=(5,5))
sns.barplot(x='Quantity',y='Sales', data=data, palette='Spectral')
plt.suptitle('Pie Consumption Patterns in the United States', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```
